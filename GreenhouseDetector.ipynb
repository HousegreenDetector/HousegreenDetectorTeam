1. Presentar una posible solución factible y versátil para reducir los efectos causados por los gases de efecto invernadero en Centroamérica. 

```python
plt.plot(GTM.year, GTM.total_ghg)
plt.plot(HND.year, HND.total_ghg)
plt.plot(SLV.year, SLV.total_ghg)
plt.plot(NIC.year, NIC.total_ghg)
plt.plot(CR.year, CR.total_ghg)
plt.plot(PAN.year, PAN.total_ghg)
plt.legend(["Guatemala", "Honduras", "El Salvador", "Nicaragua", "Costarica", "Panama"])
plt.xlabel("Años")
plt.ylabel("Cant. de emisiones de gases de efecto invernadero en CA")
plt.show()
```

2. Exponer y diferenciar los efectos de los diferentes gases (Metano, Óxido nitroso, CO2) en el ambiente con ayuda de gráficas que comparan la situación actual con la de hace 10 años, en los países de Centroamérica.

```python
# Mostrando los GEI y otros gases que han afectado en la atmosfera en El Salvador en los ultimos 10 años.
```

```python
años_2010_2020 = SLV[(SLV['year'] >= 2010) & (SLV['year']<= 2020)]
```

```python
años_2010_2020[['country','year', 'methane', 'co2','nitrous_oxide']]
```

```python
plt.plot(años_2010_2020.year, años_2010_2020.methane) 
plt.plot(años_2010_2020.year, años_2010_2020.co2)
plt.plot(años_2010_2020.year, años_2010_2020.nitrous_oxide)
plt.legend(["Metano", "CO2", "Oxido nitroso"])
plt.xlabel("Años")
plt.ylabel("Cant. de emisiones de gases en EL SALVADOR")
plt.show()
```
